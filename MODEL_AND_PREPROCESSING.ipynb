```python
import numpy as np
import pandas as pd
```

```python
df_credits = pd.read_csv("tmdb_5000_credits.csv")
df_movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
df_credits.head(1)
```

```python
df_movies.head(1)
```

```python
df = df_movies.merge(df_credits, on="title")
```

```python
df.head(1)
```

```python
df = df[['movie_id','title','overview','genres','keywords','cast','crew']]
df.head(1)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
import ast

def NameOut(LIST):
    genres = []
    for i in ast.literal_eval(LIST):
        genres.append(i["name"])
    return genres
```

```python
df.genres = df.genres.apply(NameOut)
```

```python
df.genres.head()
```

```python
df.keywords= df.keywords.apply(NameOut)
df.keywords.head()
```

```python
def LeadCast(text):
    Lead = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            Lead.append(i['name'])
        counter+=1
    return Lead
```

```python
df.cast = df.cast.apply(LeadCast)
df.cast.head()
```

```python
def fetch_director(text):
    director = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            director.append(i['name'])
    return director
```

```python
df.crew = df.crew.apply(fetch_director)
df.crew.head()
```

```python
df.head()
```

```python
def collapse(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.overview.head(1)
```

```python
df["tags"] = df['overview'] + df['genres'] + df['keywords'] + df['genres'] + df['cast'] + df['crew']
```

```python
df.head(1)
```

```python
df = df[['movie_id','title','tags']]
```

```python
df['title'] = df['title'].str.lower()
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = df[df['title'] == movie.lower()].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```
